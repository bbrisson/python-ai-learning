## 1. Architecture

Avant d'installer quoi que ce soit, il est essentiel de comprendre comment les pièces s'assemblent.

```
Navigateur / VSCode
    └── JupyterLab (serveur web — env conda base, Python 3.13)
            ├── Kernel ai_learning  (env conda ai_learning, Python 3.11)
            ├── Kernel autre_projet (env conda autre_projet, Python 3.12)
            └── ...
```

**JupyterLab** est un serveur web qui tourne en arrière-plan. Il ne fait pas tourner votre code directement — il délègue l'exécution à un **kernel**.

**Un kernel** = un processus Python séparé, dans son propre environnement conda. C'est lui qui exécute les cellules de code. JupyterLab et le kernel communiquent via le protocole ZeroMQ.

**Analogie C++ :** JupyterLab est comme un IDE. Le kernel est le compilateur/runtime. L'IDE peut appeler plusieurs compilateurs différents (MSVC, GCC, Clang) — JupyterLab peut utiliser plusieurs kernels.

**Conséquence architecturale :**
- JupyterLab s'installe **une seule fois** dans `base`
- Chaque projet a son **propre environnement conda** avec `ipykernel`
- `ipykernel` est la bibliothèque qui permet à un env conda d'être utilisé comme kernel
- Les deux sont **indépendants** : mettre à jour JupyterLab n'affecte pas les kernels, et vice versa

## 2. Installer Miniconda (mode utilisateur)

**Pourquoi Miniconda et pas Anaconda ?**
Anaconda installe ~3 Go de paquets dont la majorité ne seront jamais utilisés. Miniconda installe uniquement `base` (conda + Python), et on ajoute ce dont on a besoin.

**Pourquoi le mode utilisateur ?**
Si Miniconda est installé en mode système (`C:\ProgramData\miniconda3`), chaque `conda install` nécessite des droits administrateur. En mode utilisateur (`C:\Users\benoi\miniconda3`), on gère ses environnements librement.

### Procédure

1. Télécharger l'installeur Windows 64-bit depuis : https://docs.anaconda.com/miniconda/

2. Lancer l'installeur et configurer :
   - **Install for:** `Just Me` ← important
   - **Destination:** `C:\Users\benoi\miniconda3` (par défaut)
   - **Register as system Python:** décocher ← important

### Vérification post-installation

Ouvrir un terminal PowerShell et vérifier que le dossier existe :

````python
# À exécuter dans PowerShell (avant conda init)
# ls C:\Users\benoi\miniconda3\Scripts\conda.exe
````

## 3. Configurer conda pour PowerShell

**Pourquoi cette étape ?**

`conda activate` ne fonctionne pas comme une commande ordinaire — c'est une **fonction shell** qui modifie l'environnement du processus courant (variables PATH, CONDA_PREFIX, etc.). Cette fonction doit être injectée dans le shell au démarrage via un script d'initialisation.

`conda init powershell` fait exactement ça : il modifie le profil PowerShell (`$PROFILE`) pour charger cette fonction à chaque démarrage.

**Analogie C++ :** c'est l'équivalent de `source activate.sh` dans un build system CMake/vcpkg — la commande modifie l'environnement du shell courant, pas juste un sous-processus.

Utiliser le chemin absolu car conda n'est pas encore dans le PATH :

````python
# À exécuter dans PowerShell
# C:\Users\benoi\miniconda3\Scripts\conda.exe init powershell
# → Fermer et rouvrir VSCode complètement
````

Après redémarrage de VSCode, vérifier que conda est accessible :

````python
!conda --version
````

## 4. Créer l'environnement de travail

L'environnement de travail est défini dans `environment.yml` à la racine du repo. Ce fichier est la **source de vérité** de l'environnement — il permet de le recréer identiquement sur n'importe quelle machine.

Seul `ipykernel` est listé — c'est volontaire. Les paquets scientifiques (numpy, pandas, etc.) seront ajoutés au fur et à mesure dans les notebooks correspondants.

````python
!conda env create -f environment.yml
````

````python
!conda env list
````

## 5. Enregistrer le kernel

**Pourquoi cette étape ?**

JupyterLab ne détecte pas automatiquement tous les environnements conda — il cherche des kernels dans des emplacements spécifiques (`AppData\Roaming\jupyter\kernels\` sur Windows).

`ipykernel install` crée un dossier dans cet emplacement avec un fichier `kernel.json` qui contient le chemin vers le Python de l'environnement `ai_learning`. C'est ce fichier que JupyterLab lit pour savoir comment lancer le kernel.

**Analogie C++ :** c'est comme enregistrer un compilateur dans un IDE — vous dites à l'IDE "ce compilateur s'appelle 'ai_learning', son exécutable est ici".

Activer `ai_learning` puis enregistrer le kernel :

````python
# À exécuter dans PowerShell avec ai_learning actif
# conda activate ai_learning
# python -m ipykernel install --user --name ai_learning --display-name "Python (ai_learning)"
````

**Note :** conda 26.x crée les environnements dans `C:\Users\benoi\.conda\envs\` par défaut (pas dans `miniconda3\envs\`). C'est normal — le `kernel.json` généré pointera vers `.conda\envs\ai_learning\python.exe`.

**Après l'enregistrement, VSCode doit recharger le kernel** pour lire le nouveau `kernel.json` :
- `Ctrl+Shift+P` → `Jupyter: Restart Kernel`
- ou `Ctrl+Shift+P` → `Developer: Reload Window`

Vérifier les kernels enregistrés et le chemin Python associé :

````python
!jupyter kernelspec list
````

## 6. Installer JupyterLab

JupyterLab s'installe dans l'environnement `base` — pas dans `ai_learning`. C'est la séparation architecturale décrite en section 1.

**Ce que conda installe :** JupyterLab 4.x et toutes ses dépendances (~70 paquets : serveur web tornado, gestion de protocoles ZeroMQ, interface web, etc.).

````python
# À exécuter dans PowerShell avec base actif
# conda activate base
# conda install jupyterlab -y
````

````python
!jupyter --version
````

## 7. Lancer JupyterLab

Avec `base` actif, depuis le dossier de travail :

**Ce qui se passe :**
1. Un serveur web démarre en local (port 8888 par défaut)
2. Un token d'authentification est généré
3. Le navigateur s'ouvre automatiquement sur `http://localhost:8888/lab?token=...`

**Le terminal est occupé** tant que JupyterLab tourne. Pour l'arrêter : `Ctrl+C` deux fois dans le terminal.

````python
# À exécuter dans PowerShell
# conda activate base
# cd C:\Users\benoi\work\python-ai-learning
# jupyter lab
````

## 8. Accès via VSCode

VSCode intègre son propre moteur d'exécution de notebooks — il n'a pas besoin de JupyterLab pour faire tourner des `.ipynb`.

**Ce dont VSCode a besoin :** uniquement `ipykernel` dans l'environnement (déjà installé à l'étape 4).

### Sélectionner le kernel dans VSCode

1. Ouvrir un fichier `.ipynb` dans VSCode
2. Cliquer sur **"Select Kernel"** en haut à droite
3. Choisir **"Python Environments"**
4. Sélectionner **Python (ai_learning) (Python 3.11.x)**

**Si le kernel n'apparaît pas :** faire `Ctrl+Shift+P` → `Developer: Reload Window`

### VSCode vs navigateur — quand utiliser lequel ?

| | VSCode | Navigateur (JupyterLab) |
|--|--------|------------------------|
| **Autocomplétion** | Excellente (Pylance/IntelliSense) | Basique |
| **Débogage** | Oui (breakpoints) | Non |
| **Extensions** | Riche écosystème | Extensions Jupyter spécifiques |
| **Visualisations interactives** | Parfois limité | Meilleur support |

Pour du développement Python général et du débogage → VSCode. Pour de l'exploration de données et des visualisations interactives → navigateur.
