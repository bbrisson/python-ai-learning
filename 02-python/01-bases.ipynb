# Python de base — pour programmeurs C/C++

Ce notebook couvre les fondements de Python en partant du point de vue d'un programmeur C/C++.  
Chaque concept est mis en parallèle avec son équivalent C/C++ pour accélérer la compréhension.

## 1. Variables et types

### En C/C++
```cpp
int x = 10;
double pi = 3.14;
bool flag = true;
std::string name = "Benoit";
```
Le type est déclaré explicitement et lié à la variable pour toute sa durée de vie.

### En Python
Le type est lié à **l'objet**, pas à la variable. La variable n'est qu'une **référence** vers un objet en mémoire — exactement comme un pointeur, mais géré automatiquement.

```python
x = 10        # x pointe vers un objet int
x = "Benoit"  # x pointe maintenant vers un objet str — complètement valide
```

C'est le **typage dynamique** : le type est déterminé à l'exécution, pas à la compilation.

````python
x = 10
print(f"x = {x}, type = {type(x)}")

pi = 3.14
print(f"pi = {pi}, type = {type(pi)}")

flag = True
print(f"flag = {flag}, type = {type(flag)}")

name = "Benoit"
print(f"name = {name}, type = {type(name)}")
````

### Tout est un objet

En C++, `int` est un type primitif. En Python, **même un entier est une instance de classe**.  
La fonction `type()` retourne la classe de l'objet — l'équivalent de `typeid()` en C++.

On peut vérifier l'identifiant unique de l'objet en mémoire avec `id()` — analogue à l'adresse d'un pointeur :

````python
x = 10
print(f"id de x : {id(x)}")  # adresse mémoire de l'objet

y = x  # y pointe vers le MÊME objet que x
print(f"id de y : {id(y)}")  # même adresse

y = 20  # y pointe maintenant vers un NOUVEL objet
print(f"id de y après y=20 : {id(y)}")  # adresse différente
print(f"x est toujours : {x}")  # x n'a pas changé
print(f"id de x : {id(x)}")  # adresse mémoire de l'objet
````

## 2. Structures de contrôle

### En C/C++
```cpp
if (x > 0) {
    printf("positif\n");
} else {
    printf("négatif\n");
}
```

### En Python
**Pas d'accolades.** L'indentation définit les blocs — c'est une contrainte syntaxique, pas une convention.  
**Pas de parenthèses** autour de la condition (mais elles sont tolérées).

````python
x = 10

if x > 0:
    print("positif")
elif x == 0:
    print("zéro")
else:
    print("négatif")
````

### Boucles

**`while`** — identique à C/C++ mais sans parenthèses ni accolades :

````python
i = 0
while i < 3:
    print(f"i = {i}")
    i += 1
````

**`for`** — très différent du C/C++. En Python, `for` itère toujours sur une **séquence**, jamais sur un compteur.

```cpp
// C++ : boucle sur un compteur
for (int i = 0; i < 5; i++) { ... }
```

```python
# Python : itération sur une séquence
for i in range(5): ...
```

`range(5)` génère la séquence `[0, 1, 2, 3, 4]` — c'est un objet itérable, pas un simple entier.

````python
# Équivalent de : for (int i = 0; i < 5; i++)
for i in range(5):
    print(i)

print("---")

# Équivalent de : for (int i = 2; i < 8; i += 2)
for i in range(2, 8, 2):  # range(début, fin, pas)
    print(i)
````

## 3. Fonctions

### En C/C++
```cpp
int addition(int a, int b) {
    return a + b;
}
```

### En Python
- Mot-clé `def` à la place du type de retour
- Pas de types déclarés (mais on peut les annoter — on verra ça plus tard)
- Indentation à la place des accolades

````python
def addition(a, b):
    return a + b

print(addition(3, 4))
print(addition("Hello ", "World"))  # fonctionne aussi avec des strings !
````

### Valeurs par défaut et arguments nommés

Python supporte les **arguments avec valeur par défaut** (comme C++ depuis C++11)  
et les **arguments nommés** — une fonctionnalité absente de C.

````python
def saluer(nom, greeting="Bonjour"):
    print(f"{greeting}, {nom}!")

saluer("Benoit")                    # utilise la valeur par défaut
saluer("Benoit", "Salut")          # remplace la valeur par défaut
saluer(greeting="Hey", nom="Benoit")  # arguments nommés, ordre libre
````

## 4. Structures de données principales

### Liste — l'équivalent du tableau dynamique (`std::vector`)

- Taille variable
- Peut contenir des types mixtes (contrairement à `std::vector<T>`)
- Indexation à partir de 0 (comme C/C++)

````python
ma_liste = [10, 20, 30, 40, 50]

print(ma_liste[0])   # premier élément : 10
print(ma_liste[-1])  # dernier élément : 50 (index négatif = depuis la fin)
print(ma_liste[1:3]) # slicing : éléments index 1 et 2 → [20, 30]

ma_liste.append(60)  # équivalent de push_back()
print(ma_liste)
````

### Tuple — liste immuable

Comme une liste, mais **non modifiable après création** — analogue à un tableau `const` en C++.  
Utile pour des données qui ne doivent pas changer (coordonnées, configuration, etc.).

````python
coordonnees = (48.8566, 2.3522)  # latitude, longitude Paris

print(coordonnees[0])  # lecture OK

# coordonnees[0] = 0  # → TypeError : un tuple est immuable
````

### Dictionnaire — l'équivalent de `std::map` ou d'une struct

Paires clé→valeur. Les clés peuvent être n'importe quel type immuable (string, int, tuple...).

````python
personne = {
    "nom": "Benoit",
    "age": 40,
    "langages": ["C", "C++", "LabVIEW", "Python"]
}

print(personne["nom"])       # accès par clé
print(personne["langages"])  # la valeur peut être une liste

personne["ville"] = "Montréal"  # ajout d'une nouvelle clé
print(personne)
````
